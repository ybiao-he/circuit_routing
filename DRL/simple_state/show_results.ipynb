```python
# read rewards from csv file and plot
import numpy as np

def read_rew(filename):
    rews = np.genfromtxt(filename, delimiter=',')
    return rews

rews = read_rew("./ave_rew.csv")

import matplotlib.pyplot as plt
plt.plot(rews)
plt.xlabel('epochs')
plt.ylabel('average rewards')
plt.show()
```
